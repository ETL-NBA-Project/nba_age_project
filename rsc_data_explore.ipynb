```python
import pandas as pd
```

```python
df = pd.read_csv('final_frame.csv')
```

```python
df.head()
```

```python
import re

z_score_cols = []

for col in df.columns:
    match = re.search("_Z.*core",col)
    
    if match:
        print(match.string)
        z_score_cols.append(match.string)
    else:
        print("no match")
        
print(z_score_cols)
```

```python
# Get only Z score columns and player info.
zscore_df = df[['Player Name', 'Season Start', 'Age','Team']+z_score_cols]
```

```python
zscore_df.head()
```

```python
import numpy as np

for row in zscore_df.itertuples():
    print(row)
    print("-"*20)
    print()
    break
    
# Calculate mean for all z scores except for G_zscore and MP_zscore. 7:17 are all the zscores exclding G and MP.
df['Zscore_avg'] = [np.mean(row[7:17]) for row in zscore_df.itertuples()]
```

```python
df.head()
```

```python
# convert df to csv:
df.to_csv("final_frame.csv",index=False)
```

```python
dic = {}

for col in df.columns:
    dic[col] = col.replace(" ","_").lower()

df = df.rename(columns=dic)
```

```python
df.to_csv("final_frame.csv",index=False)
```
