```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from sqlalchemy.orm import Session
```

```python
pow_csv = "NBA_player_of_the_week.csv"
player_week_df = pd.read_csv(pow_csv, error_bad_lines=False)
player_week_df.head()
```

```python
player_list = player_week_df["Player"].tolist()
player_list
```

```python
stats_csv = "NBA Players - Basic Season Stats (1950-2017).csv"
player_stats_df = pd.read_excel(stats_csv)
player_stats_df.head()



```

```python
player_stats_df = player_stats_df[player_stats_df['Player Name'].isin(player_list)]
player_stats_df.head()
```

```python

player_stats_df.columns
# player_stats_df
# print(player_stats_df['AST%'])

# (Tm, Season Start, Age, Player Name, Pos, MP, PER, FG%, FT%, TS%, PTS, ORB, DRB, STL, BLK, TOV)
```

```python
new_stats_df = player_stats_df[['Tm', 'Season Start', 'Age', 
                               'Player Name', 'Pos', 'G', 'MP', 'PER', 
                               'FG%', 'FT%', 'TS%', 'PTS', 'ORB', 
                               'DRB', 'STL', 'BLK', 'TOV']].reset_index()
new_stats_df = new_stats_df.drop('index', axis = 1)


new_stats_df.head()
```

```python
new_stats_df['PPG'] = round(new_stats_df['PTS']/new_stats_df["G"], 2)
new_stats_df['ORB/G'] = round(new_stats_df['ORB']/new_stats_df["G"], 2)
new_stats_df['DRB/G'] = round(new_stats_df['DRB']/new_stats_df["G"], 2)
new_stats_df['SPG'] = round(new_stats_df['STL']/new_stats_df["G"], 2)
new_stats_df['BPG'] = round(new_stats_df['BLK']/new_stats_df["G"], 2)
new_stats_df['TOV/G'] = round(new_stats_df['TOV']/new_stats_df["G"], 2)

final_df = new_stats_df[['Tm', 'Season Start', 'Age', 
                        'Player Name', 'Pos', 'G', 
                        'MP', 'PER', 'FG%', 'FT%', 'TS%',
                        'PPG', 'ORB/G', 'DRB/G', 'SPG',
                        'BPG', 'TOV/G']]
final_df = final_df.sort_values(['Player Name','Season Start'])
final_df = final_df.reset_index()
final_df = final_df.drop('index', axis = 1)

final_df.head()
```
